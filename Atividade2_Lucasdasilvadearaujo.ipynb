```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Carregar o arquivo .xls
# Make sure the file is uploaded and the path is correct.
file_path = "/content/default_of_credit_card_clients__courseware_version_1_21_19.xls"
df = pd.read_excel(file_path, header=0)  # A planilha começa na linha 1 (índice 0)

# Visualizar as primeiras linhas
df.head()
```

```python
bill_features = ['BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6']
pay_features = ['PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']

print("Características de fatura:", bill_features)
print("Características de pagamento:", pay_features)
```

```python
df[bill_features].describe()
```

```python
df[bill_features].hist(bins=20, figsize=(12,8))
plt.suptitle('Distribuição dos Valores das Faturas (BILL_AMT)')
plt.show()
```

```python
df[pay_features].describe()
```

```python
df[pay_features].hist(bins=20, figsize=(12,8), xrot=45)
plt.suptitle('Distribuição dos Pagamentos (PAY_AMT)')
plt.show()
```

```python
mask_zero = (df[pay_features] == 0)
zero_counts = mask_zero.sum()
print(zero_counts)
```

```python
# Ignorar valores 0
df_nonzero = df[pay_features][df[pay_features] > 0]

# Aplicar logaritmo
df_log = df_nonzero.apply(np.log10)

# Plotar histogramas
df_log.hist(bins=20, figsize=(12,8))
plt.suptitle('Pagamentos (PAY_AMT) com Transformação Logarítmica')
plt.show()
```
